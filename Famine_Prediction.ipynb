# Production ARIMA

```python
from pyramid.arima import auto_arima
```

```python
dict_production_country={}
for index,rows in prod_sum.iterrows():
    country=list(rows)[0]
    temp = list(rows)[1:]
    temp.reverse()
    add = []
    for elem in temp:
        if elem == (0.0):
            break
        else:
            add.append(elem)
    add.reverse()
    dict_production_country[country] = add
```

```python
len(dict_production_country.keys())
for keys in dict_production_country.keys():
    if len(dict_production_country[keys]) <= 10:
        print(keys,len(dict_production_country[keys]))
```

```python
dict_production_country.pop('MNE', None)
dict_production_country.pop('SDN', None)
dict_production_country.pop('SRB', None)
```

```python
import time
start=time.time()
production={}
for i in dict_production_country.keys():
    temp_list = dict_production_country[i]
    temp_df = pd.DataFrame(temp_list)
    stepwise_model = auto_arima(temp_df, start_p=1, start_q=1,
                           max_p=5, max_q=5, m=1,
                           seasonal=False,
                           d=2, trace=False,
                           error_action='ignore',  
                           suppress_warnings=True, 
                           stepwise=True)
    stepwise_model.fit(temp_df)
    future_forecast = stepwise_model.predict(n_periods=17)
    production[i]=future_forecast[-1]
end=time.time()
print(end-start)
```

```python
import pickle

infile = open("production_dict.pickle",'rb')
production_dict = pickle.load(infile)

infile = open("population_dict.pickle",'rb')
population_dict = pickle.load(infile)

infile = open("gdp_dict.pickle",'rb')
gdp_dict = pickle.load(infile)

len(production_dict.keys())

len(population_dict.keys())

len(gdp_dict.keys())
```

```python
#making the the dictionary to create the final test data frame
final_merged_dict={}
for country in production_dict.keys():
    if country in gdp_dict.keys() & population_dict.keys():
        prod_pop_gdp = []
        production = production_dict[country]
        population = population_dict[country]
        gdp = gdp_dict[country]
        prod_pop_gdp.append(production)
        prod_pop_gdp.append(population)
        prod_pop_gdp.append(gdp)
        final_merged_dict[country] = prod_pop_gdp
```

```python
import pandas as pd

df = pd.DataFrame(columns=['Country','production','population','gdp','Famine'])

df

for country in final_merged_dict.keys():
    prod = final_merged_dict[country][0]
    pop = final_merged_dict[country][1]
    gdp = final_merged_dict[country][2]
    df = df.append({'Country': country, 'production': prod, 'population': pop,'gdp':gdp,'Famine':0}, ignore_index=True)

df.head()

df.to_csv("test.csv")
```

## Generating the target which if the country achieves at 2030,it might prevent the food crisis

```python
from scipy.stats import pearsonr
```

```python
famine_safe_zones=[]
for index,row in df_raw.iterrows():
    temp=[]
    if row['famine'] == 0:
        prod = row['production']
        pop = row['population']
        gdp = row['gdp']
        temp.append(prod)
        temp.append(pop)
        temp.append(gdp)
        famine_safe_zones.append(temp)
        temp=[]
```

```python
test_fine = test_data.drop("Unnamed: 0",axis=1)
```

```python
current_state=[]
famine_code = famine_risk[0]#select teh first country running the risk of a famine
for index,row in test_fine.iterrows():
    temp=[]
    if row['Country'] == famine_code:
        #extract features
        prod = row['production']
        pop = row['population']
        gdp = row['gdp']
        temp.append(prod)
        temp.append(pop)
        temp.append(gdp)
        current_state.append(temp) #the current state of the country which is slowly leading towards a food-cris
        break
```

```python
distance=[]
target_to_achieve = [] #the minimum target values if achieved by the country, might prevent a food crisis
cor,p = pearsonr(famine_safe_zones[0], current_state[0])
distance.append(cor)
for country in famine_safe_zones:
    corr,p_value = pearsonr(country, current_state[0])
    if corr < min(distance):
        target_to_achieve = country
        distance.append(corr)
```
